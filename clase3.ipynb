```python
import pandas as pd
import matplotlib.pyplot as plt

# Creamos un DataFrame de ejemplo con pandas
datos = {
    "Día": ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"],
    "Ventas": [10, 20, 15, 25, 30]
}
df = pd.DataFrame(datos)

# Mostramos el DataFrame en consola
print("DataFrame de prueba:")
print(df)

# Graficamos las ventas
plt.plot(df["Día"], df["Ventas"], marker="o", linestyle="-", color="blue")
plt.title("Ventas semanales")
plt.xlabel("Día")
plt.ylabel("Cantidad")
plt.grid(True)

# Mostramos el gráfico en una ventana
plt.show()
```
